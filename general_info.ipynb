```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# load data set using pandas
# we have 3 datasets load matches.csv
df=pd.read_csv("C:\EDA PROJECT MCA\matches .csv")

print(df)

# get basic info about dataset

print(df.info())
print(df.describe())

# most player of the match award

pom_counts=df["player_of_match"].value_counts()

print(pom_counts)

most_pom_player = pom_counts.idxmax()
most_pom_count = pom_counts.max()
print("The player with the most 'Man of the Match', awards is: ",most_pom_player ,"with ",most_pom_count ,"times.")

plt.figure(figsize=(10, 6))
sns.barplot(x=pom_counts.values, y=pom_counts.index, palette='viridis')
plt.title('Top Player of the Match Winners - ICC World Cup 2023')
plt.xlabel('Number of Awards')
plt.ylabel('Player Name')
plt.tight_layout()
```

```python
toss_win=df["toss_winner"].value_counts()

print(toss_win)


plt.figure(figsize=(10, 6))
sns.barplot(x=toss_win.values, y=toss_win.index, palette='viridis')
plt.title('Most toss winner  - ICC World Cup 2023')
plt.xlabel('toss wins')
plt.ylabel('team name')
plt.tight_layout()
```

```python
# Fill NaN with 0 for easy comparison
df['winner_runs'] = df['winner_runs'].fillna(0)
df['winner_wickets'] = df['winner_wickets'].fillna(0)


bat_first_wins = df[df['winner_runs'] > 0].shape[0]
bowl_first_wins = df[df['winner_wickets'] > 0].shape[0] 


print(f"--- Match Outcome Analysis ---")
print(f"Matches Won by Batting First (Defending): {bat_first_wins}")
print(f"Matches Won by Bowling First (Chasing):  {bowl_first_wins}")



```
