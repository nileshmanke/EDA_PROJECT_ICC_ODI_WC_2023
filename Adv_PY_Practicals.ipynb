```python
class Parent:
    def greet(self):
        print("greeting from parent")

class Child(Parent):
    def greet(self):
        print("gretting from child")


p1=Parent()
c1=Child()

p1.greet()
c1.greet()
    
```

```python
class Marks:
    def __init__(self,marks):
        self.marks=marks
    def __add__(self,other):
        return self.marks +other.marks

    def __str__(self):
        return f"marks={self.marks}"

m1=Marks(30)
m2=Marks(40)

print(m1+m2)


        
```

```python
class Emp:
    Company_name="TCS"
    Company_emp_count=0

    def __init__(self,name):
        self.name=name
        Emp.Company_emp_count+=1

    @classmethod

    def show_emp_count(cls):
        print("total employee",cls.Company_emp_count)
    @staticmethod
    def show_company_info():
        print("company name",Emp.Company_name)

emp1=Emp("nilesh")
emp2=Emp("abhi")

Emp.show_emp_count()
Emp.show_company_info()
         
```

```python
import pandas as pd
import numpy as np

data={
"Name":["Rahul","Sneha","Amit","Priya","Karan"],
"Age":[20,np.nan,19,22,21],
"Marks":[85,90,np.nan,88,76]
}

df=pd.DataFrame(data)

print(df)

df["Age"].fillna(df["Age"].mean(),inplace=True)
df["Marks"].fillna(df["Marks"].mean(),inplace=True)

print(df)

sorted_df=df.sort_values(by="Marks",ascending=False)

print(sorted_df)

df["Rank"]=df["Marks"].rank(ascending=False)
print(df)

print(df.describe())
```

```python
# Writing into file
file = open("sample.txt", "w")
file.write("Welcome to Python File Handling.\n")
file.write("Second Line.\n")
file.write("Third Line.\n")
file.close()

# Using read()
file = open("sample.txt", "r")
print("Using read():")
print(file.read())
file.close()

# Using readline()
file = open("sample.txt", "r")
print("Using readline():")
print(file.readline())
print(file.readline())
file.close()

# Using readlines()
file = open("sample.txt", "r")
print("Using readlines():")
print(file.readlines())
file.close()
```

```python
file = open("pointer_demo.txt", "w+")
file.write("Python File Pointer Example")

file.seek(0)
print("Initial Position:", file.tell())

data = file.read(6)
print("Read Data:", data)

print("Position after reading:", file.tell())

file.seek(7)

print("Pointer moved to:", file.tell())

print("Remaining Data:", file.read())

file.close()

# Key Concepts
# seek(n) → move pointer to position n
# tell() → get current position
# read(n) → read n characters
```
