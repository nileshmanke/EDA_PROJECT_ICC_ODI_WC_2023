```python
# import python libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns

# Imported required Python libraries.


# NumPy → Numerical calculations.
# Pandas → Data handling & cleaning.
# Matplotlib & Seaborn → Data visualization.

#  Exploratory Data Analysis (EDA) on the ICC Men's ODI World Cup 2023.



# import csv file
# load data set using pandas
# we have 3 datasets load all these 3 datasets

matches=pd.read_csv(r"C:\EDA PROJECT MCA\matches .csv")
deliveries=pd.read_csv(r"C:\EDA PROJECT MCA\deliveries.csv")
points=pd.read_csv(r"C:\EDA PROJECT MCA\points_table.csv")

# basic cleaning/preprocessing for matches dataset

# Basic Data Understanding

print("Initial shape:", matches.shape)
print("First Five rows :",matches.head())
matches.drop_duplicates(inplace=True)
print("Shape after dropping duplicates:", matches.shape)

#check for null values

print("Null values per column:\n", matches.isnull().sum())

print(matches)

# get basic info about dataset
print(matches.columns)
print(matches.info())

# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
print(matches.describe())


```

```python
# basic cleaning/preprocessing for deliveries dataset

# Basic Data Understanding

print("Initial shape:", deliveries.shape)
print("First Five rows :",deliveries.head())
deliveries.drop_duplicates(inplace=True)
print("Shape after dropping duplicates:", deliveries.shape)

#check for null values

print("Null values per column:\n", deliveries.isnull().sum())

print(deliveries)

# get basic info about dataset
print(deliveries.columns)
print(deliveries.info())

# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
print(deliveries.describe())
```

```python
# Exploratory Data Analysis (EDA)

# tournament stats (Total runs, Wickets, Boundaries)

total_runs = deliveries["runs_off_bat"].sum() + deliveries["extras"].sum()

print("Total Runs Scored In Tournament: ", total_runs, "Out Of which ",deliveries["extras"].sum()," Runs Are Extra")

bat_runs = deliveries["runs_off_bat"].sum()
extras = deliveries["extras"].sum()

plt.pie([bat_runs, extras],
        labels=["Runs from Bat", "Extras"],
        autopct="%1.1f%%")

plt.title("Runs Distribution")
plt.show()


# Total Wickets In Tournament

total_wickets = deliveries["wicket_type"].notna().sum()

print("Total Wickets In Tournament :", total_wickets)


wicket_counts = deliveries["wicket_type"].value_counts()

print(wicket_counts)


plt.figure(figsize=(10,6))

plt.bar(wicket_counts.index, wicket_counts.values)

plt.title("Wicket Types Distribution")
plt.xlabel("Wicket Type")
plt.ylabel("Number of Wickets")

plt.xticks(rotation=45)

plt.show()

print("This Plot Shows That Maximum Wickets Falls Through Caught.")

#total boundaries in tournament

# count fours
fours = (deliveries["runs_off_bat"] == 4).sum()

# count sixes
sixes = (deliveries["runs_off_bat"] == 6).sum()

# total boundaries
total_boundaries = fours + sixes

print("Total Fours:", fours)
print("Total Sixes:", sixes)
print("Total Boundaries:", total_boundaries)

labels = ["Fours", "Sixes"]
values = [fours, sixes]

plt.figure(figsize=(7,7))

plt.pie(values, labels=labels, autopct="%1.1f%%", startangle=90)

plt.title("Boundaries Distribution")

plt.show()

print("Pie chart will show 4s dominate over 6s")
```

```python
# calculate total runs per batsman
top_batsmen = deliveries.groupby("striker")["runs_off_bat"].sum()

print(top_batsmen)

# sort and get top 10
top10 = top_batsmen.sort_values(ascending=False).head(10)

print(top10)

plt.figure(figsize=(10,6))

plt.bar(top10.index, top10.values)

plt.title("Top 10 Run Scorers - ICC World Cup 2023")
plt.xlabel("Batsman")
plt.ylabel("Total Runs")

plt.xticks(rotation=45)

plt.show()

print("The analysis shows the top performing batsmen of the tournament. Players like Virat Kohli and Rohit Sharma dominated the run charts, indicating strong batting performance during the World Cup.")



#Top 10 Wicket Takers

# filter rows where wicket occurred
wickets = deliveries[(deliveries["wicket_type"].notna()) & (deliveries["wicket_type"] != "run out")]

# count wickets by bowler
top_bowlers = wickets.groupby("bowler").size()

# sort and get top 10
top10_bowlers = top_bowlers.sort_values(ascending=False).head(10)

print(top10_bowlers)


plt.figure(figsize=(10,6))

plt.bar(top10_bowlers.index, top10_bowlers.values)

plt.title("Top 10 Highest Wicket Takers - ICC World Cup 2023")
plt.xlabel("Bowler")
plt.ylabel("Wickets")

plt.xticks(rotation=45)

plt.show()

print("The analysis identifies the most successful bowlers of the ICC World Cup 2023. Bowlers like Mohammed Shami and Adam Zampa dominated the tournament with the highest wicket counts, showing their major impact on match outcomes.")
```

```python
#Best Striker of the Ball (Top Batsmen by Strike Rate)

#strike_rate = (runs / balls) * 100

import pandas as pd

runs = deliveries.groupby("striker")["runs_off_bat"].sum()
balls = deliveries.groupby("striker").size()

strike_rate = (runs / balls) * 100

sr_df = pd.DataFrame({
    "Runs": runs,
    "Balls": balls,
    "Strike Rate": strike_rate
})

print("Sometimes players who faced very few balls get very high strike rate So filter players who faced at least 100 balls.")
sr_df = sr_df[sr_df["Balls"] >= 100]
top_sr = sr_df.sort_values("Strike Rate", ascending=False).head(10)

print(top_sr)



top_sr["Strike Rate"].plot(kind="bar", figsize=(10,6))

plt.title("Top 10 Batsmen by Strike Rate")
plt.xlabel("Batsman")
plt.ylabel("Strike Rate")

plt.show()
```

```python
deliveries["over"] = deliveries["ball"].astype(int)

def phase(over):
    if over <= 10:
        return "Powerplay"
    elif over <= 40:
        return "Middle Overs"
    else:
        return "Death Overs"

deliveries["phase"] = deliveries["over"].apply(phase)


runs_phase = deliveries.groupby("phase")["runs_off_bat"].sum()

print(runs_phase)


plt.figure(figsize=(8,6))

plt.bar(runs_phase.index, runs_phase.values)

plt.title("Run Distribution Across Match Phases")
plt.xlabel("Match Phase")
plt.ylabel("Runs")

plt.show()

wickets_phase = deliveries[deliveries["wicket_type"].notna()].groupby("phase").size()
print(wickets_phase)

wickets_phase.plot(kind="bar")

plt.title("Wickets in Each Phase")
plt.xlabel("Match Phase")
plt.ylabel("wickets")

plt.show()


balls_phase = deliveries.groupby("phase").size()
print(balls_phase)

overs_phase = balls_phase / 6

run_rate = runs_phase / overs_phase

print(run_rate)

run_rate.plot(kind="bar", figsize=(8,6))

plt.title("Run Rate by Match Phase")
plt.xlabel("Match Phase")
plt.ylabel("Run Rate")

plt.show()



print("The phase-wise analysis shows how teams scored runs during different stages of the match. Most runs were scored during the middle overs, where batsmen focused on building partnerships. The death overs phase showed aggressive batting with higher scoring rates, while the powerplay provided early scoring opportunities due to field restrictions.")
print("Run rate analysis shows that scoring acceleration occurs during the Death Overs, where batsmen play aggressively to maximize runs. The Powerplay provides a steady start due to field restrictions, while the Middle Overs focus on strike rotation and partnership building. Therefore, teams that maintained a high run rate during the middle and death overs had a higher probability of winning matches.")
```

```python
# check if toss winner also won the match
matches["toss_win_match_win"] = matches["toss_winner"] == matches["winner"]

print(matches[["toss_winner","winner","toss_win_match_win"]])


toss_result = matches["toss_win_match_win"].value_counts()

print(toss_result)



toss_result.plot(kind="bar")

plt.title("Impact of Toss on Match Result")
plt.xlabel("Toss Winner Also Won Match")
plt.ylabel("Number of Matches")

plt.xticks([0,1], ["Lost After Winning Toss","Won After Winning Toss"], rotation=0)

plt.show()

print("The analysis shows that winning the toss did not provide a significant advantage in the ICC Men's ODI World Cup 2023. In fact, teams that won the toss lost more matches (60.4%) than they won (39.6%). This indicates that match outcomes were more strongly influenced by team performance, batting strength, bowling strategy, and match conditions, rather than the toss decision itself.")
```

```python
# Venue Stats: Pitch behavior across different stadiums in India.

import pandas as pd

df = pd.read_csv("deliveries.csv")

df["total_runs"] = df["runs_off_bat"] + df["extras"]

venue_runs = df.groupby("venue")["total_runs"].sum()

print(venue_runs.sort_values(ascending=False))



venue_runs.sort_values(ascending=False).plot(kind="bar", figsize=(12,6))

plt.title("Total Runs by Venue")
plt.xlabel("Venue")
plt.ylabel("Runs")

plt.show()

print("Wankhede Stadium recorded the highest total runs, indicating a batting-friendly pitch. The ground is known for short boundaries and good bounce, which helps batsmen score freely.")


venue_wickets = df[df["wicket_type"].notna()].groupby("venue").size()

print(venue_wickets.sort_values(ascending=False))


venue_wickets.sort_values(ascending=False).plot(kind="bar", figsize=(12,6))

plt.title("Wickets by Venue")
plt.xlabel("Venue")
plt.ylabel("Number of Wickets")

plt.show()

print("Despite high runs, Dharamsala also produced the highest number of wickets, indicating a balanced pitch where both batsmen and bowlers had opportunities.")


print("The venue analysis shows that pitch behavior varied significantly across stadiums. Some venues like Wankhede and Delhi favored batting, while others like Dharamsala offered a balanced contest between bat and ball. These variations highlight how pitch conditions and ground dimensions influence match strategies and outcomes.")
```
